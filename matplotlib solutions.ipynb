```python
def make_z(x,y):
    z = np.zeros([len(x),len(y)])
    for i in range(len(x)):
        for j in range(len(y)):
            z[j][i] = 2*x[i]**2 + y[j]**4
            
    return z

x = np.linspace(0,10,100)
y = np.linspace(0,10,100)
z = make_z(x,y)
```

```python
plt.contour(x,y,z,1000)
plt.colorbar()
```
